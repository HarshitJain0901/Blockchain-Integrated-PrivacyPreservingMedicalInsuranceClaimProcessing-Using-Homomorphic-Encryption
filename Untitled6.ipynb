```python
#Removal of outlier:

df4 = df3.copy()

for i in [i for i in df4.columns]:
    if df4[i].nunique()>=12:
        Q1 = df4[i].quantile(0.25)
        Q3 = df4[i].quantile(0.75)
        IQR = Q3 - Q1
        df4 = df4[df4[i] <= (Q3+(1.5*IQR))]
        df4 = df4[df4[i] >= (Q1-(1.5*IQR))]
df4 = df4.reset_index(drop=True)
display(df4.head())
print('\n\033[1mInference:\033[0m Before removal of outliers, The dataset had {} samples.'.format(df1.shape[0]))
print('\033[1mInference:\033[0m After removal of outliers, The dataset now has {} samples.'.format(df4.shape[0]))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm  # Handles notebooks and scripts
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Initialize containers
DROP = []
scores1, scores2, scores3 = [], [], []

# Baseline models with all features
scores1.append(
    f1_score(Test_Y, LogisticRegression(max_iter=1000).fit(Train_X_std, Train_Y).predict(Test_X_std), average='weighted') * 100
)
scores2.append(
    f1_score(Test_Y, RandomForestClassifier().fit(Train_X_std, Train_Y).predict(Test_X_std), average='weighted') * 100
)
scores3.append(
    f1_score(Test_Y, XGBClassifier(eval_metric='logloss').fit(Train_X_std, Train_Y).predict(Test_X_std), average='weighted') * 100
)

# Iteratively drop features with high VIF
for _ in tqdm(range(len(Train_X_std.columns) - 1)):
    # Compute VIFs on remaining features
    Xs = Train_X_std.drop(DROP, axis=1)
    vif = pd.DataFrame()
    vif["Features"] = Xs.columns
    vif["VIF"] = [variance_inflation_factor(Xs.values, j) for j in range(Xs.shape[1])]
    vif = vif.sort_values(by="VIF", ascending=False).reset_index(drop=True)

    # Stop if all VIFs are acceptable
    if vif.VIF.iloc[0] <= 5:
        break

    # Drop most collinear feature
    DROP.append(vif.Features.iloc[0])

    # Update datasets
    Train_X_drop = Train_X_std.drop(DROP, axis=1)
    Test_X_drop = Test_X_std.drop(DROP, axis=1)

    # Fit models and store scores
    scores1.append(
        f1_score(Test_Y, LogisticRegression(max_iter=1000).fit(Train_X_drop, Train_Y).predict(Test_X_drop), average='weighted') * 100
    )
    scores2.append(
        f1_score(Test_Y, RandomForestClassifier().fit(Train_X_drop, Train_Y).predict(Test_X_drop), average='weighted') * 100
    )
    scores3.append(
        f1_score(Test_Y, XGBClassifier(eval_metric='logloss').fit(Train_X_drop, Train_Y).predict(Test_X_drop), average='weighted') * 100
    )

# Plot F1 score changes
plt.figure(figsize=(10, 6))
plt.plot(scores1, label='Logistic Regression', marker='o')
plt.plot(scores2, label='Random Forest', marker='s')
plt.plot(scores3, label='XGBoost', marker='^')
plt.xlabel('Number of Features Dropped')
plt.ylabel('Weighted F1 Score (%)')
plt.title('Model Performance vs Feature Reduction (VIF-based)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Optional: print the dropped features
print("\nFeatures dropped due to high multicollinearity:")
for i, f in enumerate(DROP, 1):
    print(f"{i}. {f}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
from sklearn.feature_selection import RFE
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

# Base model for RFE
LR = LogisticRegression(max_iter=1000)

# Initialize score lists
scores1, scores2, scores3 = [], [], []

# Baseline (all features)
scores1.append(f1_score(Test_Y, LR.fit(Train_X_std, Train_Y).predict(Test_X_std), average='weighted') * 100)
scores2.append(f1_score(Test_Y, RandomForestClassifier().fit(Train_X_std, Train_Y).predict(Test_X_std), average='weighted') * 100)
scores3.append(f1_score(Test_Y, XGBClassifier(eval_metric='logloss').fit(Train_X_std, Train_Y).predict(Test_X_std), average='weighted') * 100)

# Recursive Feature Elimination Loop
for i in tqdm(range(1, len(Train_X_std.columns))):  # Start from 1 to avoid zero features
    num_features = len(Train_X_std.columns) - i
    rfe = RFE(LR, n_features_to_select=num_features)
    rfe.fit(Train_X_std, Train_Y)
    
    selected_features = Train_X_std.columns[rfe.support_]

    # Logistic Regression
    scores1.append(
        f1_score(
            Test_Y,
            LogisticRegression(max_iter=1000).fit(Train_X_std[selected_features], Train_Y).predict(Test_X_std[selected_features]),
            average='weighted'
        ) * 100
    )

    # Random Forest
    scores2.append(
        f1_score(
            Test_Y,
            RandomForestClassifier().fit(Train_X_std[selected_features], Train_Y).predict(Test_X_std[selected_features]),
            average='weighted'
        ) * 100
    )

    # XGBoost
    scores3.append(
        f1_score(
            Test_Y,
            XGBClassifier(eval_metric='logloss').fit(Train_X_std[selected_features], Train_Y).predict(Test_X_std[selected_features]),
            average='weighted'
        ) * 100
    )

# Plotting the scores
plt.figure(figsize=(10, 6))
plt.plot(range(len(scores1)), scores1, marker='o', label='Logistic Regression')
plt.plot(range(len(scores2)), scores2, marker='s', label='Random Forest')
plt.plot(range(len(scores3)), scores3, marker='^', label='XGBoost')
plt.xlabel('Number of Features Dropped')
plt.ylabel('Weighted F1 Score (%)')
plt.title('Model Performance vs Recursive Feature Elimination')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Optionally print best scores
print(f"\n Max F1-Score (Logistic Regression): {max(scores1):.2f}% at {len(Train_X_std.columns) - scores1.index(max(scores1))} features")
print(f" Max F1-Score (Random Forest): {max(scores2):.2f}% at {len(Train_X_std.columns) - scores2.index(max(scores2))} features")
print(f" Max F1-Score (XGBoost): {max(scores3):.2f}% at {len(Train_X_std.columns) - scores3.index(max(scores3))} features")
```
